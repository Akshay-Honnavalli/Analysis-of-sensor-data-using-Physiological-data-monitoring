```python
quartiles = means_df.quantile([0.25, 0.5, 0.75])

# Function to apply quartiles and replace values
def apply_quartiles(row, quartiles):
    for col in row.index:
        if row[col] < quartiles.loc[0.25, col]:
            row[col] = 'low'
        elif row[col] > quartiles.loc[0.75, col]:
            row[col] = 'high'
        else:
            row[col] = 'medium'
    return row

# Apply the function to each row
df = means_df.apply(apply_quartiles, args=(quartiles,), axis=1)

print(df)
```
